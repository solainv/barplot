# Barplot:

### Definition: 
Zeigt Daten mithilfe rechteckiger Balken an, wobei die Höhe oder Lönge der Balken die Größe der Daten repräsentiert (Balöen könnten entweder Vertikal oder Horizontal orientiert sein)
## Verwendung:
- Ideal um verschiedene Kategorien oder Gruppen zu vergleichen
- eignet sich besonders gut für discrete Daten z.B. Umsatz nach Produkt
- sie eigen deutlich wie die verschiedenen Kategorien zum Ganzen beitragen.
- ermöglichen Rangfolgen wie Verkaufsprozentsatz.
- gut für leicht geordnete Daten wie meistverkauften Produkten auf dem Markt.
- Skala soll sorgfältig festgelegt

### Daten importieren
Die Daten umfassen viele Merkmale von Personnen unter anderem Geschlecht, Alter, den BMI (Body Mass Index) und die zugehörigen Versicherungsprämien(charges). Der BMI gibt Aufschluss über das Verhältnis von Gewicht zu Körpergröße, während die Versicherungsprämien die Kosten für die Krankenversicherung darstellen.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/solainv/barplot/main/insurance.csv')
df.head()
```

```python
df_grouped = df.groupby('region').agg({'charges':'mean'})
df_grouped
```

```python
ax = df_grouped.plot(kind='bar')
ax.set_ylim(10000)
```

Die Skala spielt eine Rolle wenn wir die y-achse bei 0 (standardmässig) starten lassen  (siehe unten) dann merket man kaum den Unteschied zwischen Kategorien

```python
ax = df_grouped.plot(kind='bar')
```

Wir Könne auch das Geschlechtmerkmal hinzuufügen

```python
# Pivot-Tabelle erstellen
pivot_table = df.pivot_table(index='region', columns='sex', values='charges', aggfunc='mean')
pivot_table
```

```python
colors = ['skyblue', 'lightcoral']
ax = pivot_table.plot(kind='bar', figsize=(8, 6), color=colors)
ax.set_ylim(10000)
```
